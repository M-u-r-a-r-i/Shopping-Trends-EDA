```python
data["Gender"].value_counts().plot(kind = "bar")
plt.title("Distribution of Gender")
plt.show()
```

```python
data.groupby("Category")["Purchase Amount (USD)"].mean().sort_values().plot(kind = "bar")
```

```python
data.groupby("Category")["Purchase Amount (USD)"].mean().sort_values(ascending=False).plot(kind = "bar")
```

```python
data["Subscription Status"].value_counts().plot(kind = "pie")
plt.title("Distribution of Subscription Status")
plt.show()
```

```python
data.groupby("Season")["Purchase Amount (USD)"].sum()
```

```python
data.boxplot(column = "Purchase Amount (USD)", by = "Frequency of Purchases")
plt.title("Distribution of Frequency of Purchases")
plt.show()
```

```python
data["Payment Method"].value_counts().sort_values(ascending = False).plot(kind = "bar")
```

```python
season_sales = data.groupby("Season")["Purchase Amount (USD)"].sum().reindex(["Winter","Spring","Summer","Fall"])
plt.figure(figsize=(8,5))
season_sales.plot(kind="line", marker="o", linewidth=2, color="orange")
plt.title("Seasonal Sales ")
plt.xlabel("Season")
plt.ylabel("Purchase Amount (USD)")
plt.show()
```

```python
category_sales = data.groupby("Category")["Purchase Amount (USD)"].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
category_sales.plot(kind="bar", color="skyblue")
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Purchase Amount (USD)")
plt.xticks(rotation=45)
plt.show()
```

```python
from wordcloud import WordCloud
text = " ".join(data["Item Purchased"].astype(str))
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(text)

plt.figure(figsize=(10,6))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Top 10 best-selling products")
plt.show()
```
