```python
data["Gender"].value_counts().plot(kind = "bar")
plt.title("Distribution of Gender")
plt.show()
```

```python
data.groupby("Category")["Purchase Amount (USD)"].mean().sort_values().plot(kind = "bar")
```

```python
data.groupby("Category")["Purchase Amount (USD)"].mean().sort_values(ascending=False).plot(kind = "bar")
```
